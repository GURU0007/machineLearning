```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv(r"C:\Users\guru sai reddy\Desktop\py-master\ML\5_one_hot_encoding\Exercise\carprices1.csv")
df
```

```python
dummies=pd.get_dummies(df.CarModel)
dummies
```

```python
merge=pd.concat([df,dummies],axis='columns')
merge
```

```python
final=merge.drop(['CarModel','Mercedez Benz C class'],axis='columns')
final
```

```python
model=linear_model.LinearRegression()
```

```python
X=final.drop(['SellPrice'],axis='columns')
X
```

```python
Y=final.SellPrice
Y
```

```python
model.fit(X,Y)
```
